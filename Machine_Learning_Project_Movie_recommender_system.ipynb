# **Recommender System**

Recommeder system is used to recommend something to the user based on there interest.

Recommender system is of three type:



1.   Content Based
2.   Collaborative Filtering
3.   Hybrid


**1.  Content Based:** In content based recommender system content recommends on the basis of similiarity of content like in spotify if you are listening to romantic songs you will get recommendations for romantic songs not for patriotic songs. In this tags are created and on the basis of similarity of these tags one get recommendations.

**2.  Collaborative Filtering:** In Collaborative Filtering, we tend to find similar users and recommend what similar users like. In this type of recommendation system, we donâ€™t use the features of the item to recommend it, rather we classify the users into clusters of similar types and recommend each user according to the preference of its cluster. Facebook works on recommendation system like if your friend like a post you will also get recommendation of that post.

```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv("/content/tmdb_5000_movies.csv")
credits = pd.read_csv("/content/tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)["cast"].values
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head(1)
```

```python
# genre,id,keywords,title, overwiew,cast, crew
movies = movies[['genres','id','keywords','title', 'overview','cast', 'crew']]
```

```python
movies.head()
```

# Data Preprocessing:



```python
#Handling missing data
movies.isnull().sum()
```

```python
#removing those 3 movies whose overview is not there since 3 is not very big number
movies.dropna(inplace=True)
```

```python
#checking duplicate data
movies.duplicated().sum()
```

```python
#Creating function
def convert(obj):
  L = []
  for i in ast.literal_eval(obj): #as we can see the data is list of list and in string for so it will through error so to get a list we use this function ast.literal_eval()
    L.append(i['name'])
  return L

```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
#Creating function
def convert3(obj):
  L = []
  Count = 0
  #as we can see the data is list of list and in string for so it will through error so to get a list we use this function ast.literal_eval()

  for i in ast.literal_eval(obj):
    if i != 3:
      L.append(i['name'])
      Count = Count + 1
    else:
      break
  return L

```

```python
movies['cast']= movies['cast'].apply((convert3))
```

```python
#Creating function
def fetch_director(obj):
  L = []

  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#overview column is in string form so now we need to convert that intolist of words for that we will use .split() which will do our job we are doing this because we wants to concatinate all columnns other than id and title into one which we will use as tag.
movies['overview'] = movies[ 'overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

Now we to perform  **Data Transformation**  and for that we need to remove space between two words like Sam Wortington we need to remove space between them.

We are doing this because wwhen we write Sam Wortington then it will create two different tags tag1 Sam tag2 Wortingon and when it comes to Sam Mendes then our recomendation system get confuse between these two Sam so we need t remove whitespace between them and make them one SamWorthington.

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#Now we will concatinate all and create a new column by name tags
movies['tags'] = movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew']
```

```python
movies.head()
```

Now we will create a new dataframe

```python
New_df = movies[['id','title', 'tags']]
```

```python
New_df
```

```python
#now we will convert list into string
New_df['tags'] = New_df['tags'].apply(lambda x:" ".join(x))
```

```python
New_df.head()
```

```python
New_df['tags'][0]
```

```python
#now convert everything into lowercase as suggested
New_df['tags'] = New_df['tags'].apply(lambda x:x.lower())
```

```python
New_df.head()
```

# Vectorization/TextVectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(New_df['tags']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
    return " ".join(y)
```

```python
New_df['tags'] = New_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

Now we will calculate cosine similiarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similiarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
  movie_index = New_df[New_df['title'] == movie].index[0]
  distances = similiarity[movie_index]
  movies_list = sorted(list(enumerate(distances)) , reverse=True, key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(New_df.iloc[i[0]].title)

```

```python
recommend('Batman Begins')
```
